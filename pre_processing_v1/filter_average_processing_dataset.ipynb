```python
no2 = dataset[(dataset["Variable"]=="NO2") & (dataset["Flagged as Suspect Reading"]==False)]
no2.to_csv('./dataset/no2.csv',index=False, header=True)
no2 = pd.read_csv('./dataset/no2.csv')


no2["datetime"] = pd.to_datetime(no2["Timestamp"])
no2["datetime"]

no2["year"] = no2["datetime"].dt.year
no2["month"] = no2["datetime"].dt.month
no2["day"] = no2["datetime"].dt.day
no2["hour"] = no2["datetime"].dt.hour
no2hourlymean = no2.groupby(["year","month","day","hour"]).Value.mean()

no2hourlymean.to_csv('./dataset/no2avg.csv', index = True, header = True)
no2avg = pd.read_csv('./dataset/no2avg.csv')

no2avg.rename(columns={'Value':'No2'},inplace=True)

no2avg.to_csv('./dataset/no2avg.csv',index = False)
no2avg = pd.read_csv('./dataset/no2avg.csv')
no2avg
```
