### read datasets, group by year, month, day, hour, make a mean based on hour

```python
for filename in os.listdir('../2018'):
    if filename.endswith('.csv'):
        
        
        size_filename = len(filename)
        filename_short = filename[:size_filename - 4]
        filename_short = filename_short.replace(".", "_")
        
        path_main = '../2018/'+filename
        path_avg = '../2018_avg/' + filename_short + '_avg.csv'
        
        print(filename)
        print(filename_short)
        
        print(size_filename)
        
        print(path_main)
        print(path_avg)
        
        
        value = filename_short
        #sensor_id_value = 'sensor_id_AverageSpeed'

        
        #read the orginal dataset
        dataset = pd.read_csv(path_main)

        # datetime for the timestamp on year, month, day, hour
        dataset["datetime"] = pd.to_datetime(dataset["timestamp"])
        dataset["datetime"]

        dataset["year"] = dataset["datetime"].dt.year
        dataset["month"] = dataset["datetime"].dt.month
        dataset["day"] = dataset["datetime"].dt.day
        dataset["hour"] = dataset["datetime"].dt.hour

        # make a mean from the dataset based on hour
        datasethourlymean = dataset.groupby(["year","month","day","hour"]).value.mean()

        # read the mean dataset and saved it in avg folder
        datasethourlymean.to_csv(path_avg, index = True, header = True)
        datasetavg = pd.read_csv(path_avg)

        # rename value header and replace it based on the variable
        datasetavg.rename(columns={'value': value},inplace=True)

        # rename sensor_id header and replace it based on the variable
        #datasetavg.rename(columns={'sensor_id': sensor_id_value},inplace=True)

        # save and read the processed dataset
        datasetavg.to_csv(path_avg,index = False)
        #datasetavg = pd.read_csv(path_avg)
        #datasetavg
```
